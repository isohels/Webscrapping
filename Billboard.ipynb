### WebScrapping Billboard

```python
import bs4
from urllib.request import Request, urlopen
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
#grabbing the url

url = 'https://www.billboard.com/charts/hot-100'
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})

webpage = urlopen(req).read()

pagesoup = soup(webpage,"html.parser")

```

```python
#grabbing song names

songnames = pagesoup.findAll("span",{"class":"chart-element__information__song text--truncate color--primary"})

#songnames[0].text

#grabbing rank

songrank = pagesoup.findAll("span",{"class":"chart-element__rank flex--column flex--xy-center flex--no-shrink"})
#songrank[0].span.text

#grabbing artist
songartist = pagesoup.findAll("span",{"class":"chart-element__information__artist text--truncate color--secondary"})
len(songartist)
```

```python
filename = 'billboard.csv'
f = open(filename,"w")

headers = "rank, song, artist\n"
f.write(headers)

for i in range(0,len(songnames)):
    
    #songnames = pagesoup.findAll("span",{"class":"chart-element__information__song text--truncate color--primary"
    names=songnames[i].text
    rank = songrank[i].span.text
    artist = songartist[i].text
    print("name:" + names)
    print("rank:" + rank)
    print("artist:" + artist)
    
    f.write(names + ","+ artist+ ","+ rank+"\n")
f.close()
```
